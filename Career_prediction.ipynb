```python
#This part is for Train and Test

from sklearn.model_selection import train_test_split

x = df.iloc[:,df.columns != 'Output']
y = df.iloc[:,df.columns == 'Output']

print("Here is the value of x ------------>>>>>>>>>\n")
print(x)
print("\n\n")
print("Here is the value of y ------------>>>>>>>>>\n")
print(y)
print("\n\n")
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.10)

#Here I select 10% of my dataset for testing purpose and remaining 90% of my dataset for traning purpose
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
# Here the implementation of Random Forest Algorithm and Model building

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

# Here I train the model

# Here I use Ravel function for avoiding DataConversionWarning
model.fit(xtrain, ytrain.values.ravel()) 
```

```python
# Here I predict the output

prediction = model.predict(xtest)

#print(prediction)
```

```python
from sklearn.metrics import  accuracy_score

accuracy = accuracy_score(prediction, ytest)

print("Therefore the Accuracy is : ", accuracy)
```

```python
#Graphical Representation with Confution Matrix ( Ref: W3school )
import numpy
from sklearn import metrics
actual = numpy.random.binomial(1,.9,size = 1000)
predicted = numpy.random.binomial(1,.9,size = 1000)
confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```
